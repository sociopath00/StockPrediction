- We have data from 27th August,2018 to 25th August,2020 but in reverse order


- *Spread High-Low* and *Spread Close-Open* are derived columns that represent difference between High-Low and Open-Close values respectivly for that day

```python
# Change Column names
df.columns = DATA_COLUMNS

print(df.columns)
```

```python
# convert dates column into pandas datetime and sort in ascending order
df["dates"] = pd.to_datetime(df["dates"])
df = df.sort_values(by="dates").reset_index(drop=True)
df.head()
```

```python
plt.plot(df["close"])
plt.plot(df["open"])
plt.plot(df["high"])
plt.plot(df["low"])
plt.title("Stock Prices")
plt.show()
```
